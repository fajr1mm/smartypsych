## TRAIN/TEST SPLIT & LOAD DATA

```python
file_path = 'Contoh Respon dan Penilaian Rekapitulasi_Data BEQ dan Level.xlsx'

xls = pd.ExcelFile(file_path)
sheet_names = xls.sheet_names

dfx = [pd.read_excel(file_path, sheet_name=sheet) for sheet in sheet_names[:-1]]

dfc = pd.concat(dfx, ignore_index=True)

dfc = dfc.drop(columns=[dfc.columns[0]] + list(dfc.columns[5:12]))
print(dfc.dtypes)
dfc
```

```python
# df['translated'] = df['answer'].apply(translate_text)
dfc['LEVEL KOMPETENSI'] = dfc['LEVEL KOMPETENSI'].astype(str)
dfc['RESPON'] = dfc['RESPON'].astype(str)
dfc.dtypes
```

```python
# Inisialisasi variabel untuk menyimpan hasil
result = {'A': [], 'B': [], 'C': [], 'D': []}

# Iterasi melalui setiap baris DataFrame
for i in range(len(dfc)):
    # Jika kolom A tidak null, tambahkan ke hasil baru
    if not pd.isna(dfc['INISIAL '].iloc[i]):
        result['A'].append(dfc['INISIAL '].iloc[i])
        result['B'].append(dfc['KOMPETENSI'].iloc[i])
        result['C'].append(dfc['RESPON'].iloc[i])
        result['D'].append(dfc['LEVEL KOMPETENSI'].iloc[i])
    # Jika kolom A null, gabungkan kolom C dengan baris di atasnya yang A tidak null
    else:
        result['C'][-1] += '-' + dfc['RESPON'].iloc[i]

# Buat DataFrame hasil
df = pd.DataFrame(result)
df
```

```python
# df['translated'] = df['answer'].apply(translate_text)
df['LEVEL KOMPETENSI'] = df['LEVEL KOMPETENSI'].astype(str)
```

```python
# yahoo_answers_qa = yahoo_answers_qa["train"].train_test_split(test_size=0.3)
from sklearn.model_selection import train_test_split

# Membagi dataset menjadi data train (70%) dan data test (30%)
df_train, df_test = train_test_split(df, test_size=0.3, random_state=42)

df_train
```
